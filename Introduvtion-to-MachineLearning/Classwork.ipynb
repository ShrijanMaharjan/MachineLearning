```python
import pandas as pd
import numpy as np
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/car/Customer Call List.xlsx')
df
```

```python
df_cleaned = df.drop_duplicates();
df_cleaned
```

```python
new_df=df_cleaned.drop(columns="Not_Useful_Column")
new_df

```

```python
new_df['Phone_Number'] = new_df['Phone_Number'].replace(['NaN', 'N/A','N/a'], ' ').fillna(' ')
new_df

```

```python
new_df['Last_Name'] = new_df['Last_Name'].replace({'/': ' ', r'\.\.\.': ' ', '_': ' '}, regex=True)

new_df
```

```python
def format_phone_number(phone):
    if pd.isna(phone):
        return ' '

    phone = ''.join(filter(str.isdigit, str(phone)))
    if len(phone) < 6:
        return phone
    elif len(phone) <= 10:
        return f"{phone[:3]}-{phone[3:6]}-{phone[6:]}"
    else:
        return f"{phone[:3]}-{phone[3:6]}-{phone[6:]}"

new_df['Phone_Number'] = new_df['Phone_Number'].apply(format_phone_number)
new_df
```

```python
new_df[['Street_Address', 'State', 'Zip_Code']] = new_df['Address'].str.extract(r'^(.*?)(?:,\s*([^,\d]+))?(?:,\s*(\d{5}))?$')

# Drop the original Address column if no longer needed
new_df.drop(columns=['Address'], inplace=True)
```

```python
new_df
```

```python
new_df['Do_Not_Contact'] = new_df['Do_Not_Contact'].replace({'Yes': 'Y', 'No': 'N'})
new_df
```
